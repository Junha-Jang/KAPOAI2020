```python
def all_move(A, player):
    plist = []
    for x1 in range(BOARD_ROWS):
        for y1 in range(BOARD_COLS):
            if A[x1][y1] != player:
                continue
            for x2 in range(max(x1 - 2, 0), min(x1 + 3, n)):
                for y2 in range(max(y2 - 2, 0), min(y2 + 3, n)):
                    if A[x2][y2] == "0":
                        plist.append((x1, y1, x2, y2))
    return plist
```

```python
def rand_move():
    plist = all_move(B, 1)
    if not plist:
        return (-1, -1, -1, -1)
    return plist[np.random.choice(len(plist))]
```

```python
def get_phase():
    cnt = 0
    for x in range(BOARD_ROWS):
        for y in range(BOARD_COLS):
            if B[x][y] != "0":
                cnt += 1
    if cnt >= BOARD_ROWS * BOARD_COLS - 10:
        return 3

    for x1 in range(BOARD_ROWS):
        for y1 in range(BOARD_COLS):
            if B[x1][y1] != 1:
                continue
            for x2 in range(max(x1 - 3, 0), min(x1 + 4, n)):
                for y2 in range(max(y1 - 3, 0), min(y1 + 4, n)):
                    if B[x2][y2] == 2:
                        return 2
    return 1
```

```python
def phase1(player):
    cntu = cntd = cntl = cntr = 0
    for x in range(BOARD_ROWS):
        for y in range(BOARD_COLS):
            if B[x][y] != "2" if player == "1" else "1":
                continue
            if x > y:
                if 6-x > y:
                    cntl += 1
                elif 6-x < y:
                    cntd += 1
            elif x < y:
                if 6-x > y:
                    cntu += 1
                elif 6-x < y:
                    cntr += 1

    p = ()
    mx = max(max(cntu, cntd), max(cntl, cntr))
    if mx == cntu:
        if B[1][5] == "0":
            p = (0, 6, 1, 5)
        elif B[2][4] == "0":
            p = (1, 5, 2, 4)
        elif B[1][4] == "0":
            p = (2, 4, 1, 4)
        elif B[2][5] == "0":
            p = (2, 4, 2, 5)
        elif B[5][1] == "0":
            p = (6, 0, 5, 1)
        elif B[4][2] == "0":
            p = (5, 1, 4, 2)
        elif B[4][1] == "0":
            p = (4, 2, 4, 1)
        elif B[5][2] == "0":
            p = (4, 2, 5, 2)
        else:
            p = rand_move()
    elif mx == cntd:
        if B[5][1] == "0":
            p = (6, 0, 5, 1)
        elif B[4][2] == "0":
            p = (5, 1, 4, 2)
        elif B[5][2] == "0":
            p = (4, 2, 5, 2)
        elif B[4][1] == "0":
            p = (4, 2, 4, 1)
        elif B[1][5] == "0":
            p = (0, 6, 1, 5)
        elif B[2][4] == "0":
            p = (1, 5, 2, 4)
        elif B[2][5] == "0":
            p = (2, 4, 2, 5)
        elif B[1][4] == "0":
            p = (2, 4, 1, 4)
        else:
            p = rand_move()
    elif mx == cntl:
        if B[5][1] == "0":
            p = (6, 0, 5, 1)
        elif B[4][2] == "0":
            p = (5, 1, 4, 2)
        elif B[4][1] == "0":
            p = (4, 2, 4, 1)
        elif B[5][2] == "0":
            p = (4, 2, 5, 2)
        elif B[1][5] == "0":
            p = (0, 6, 1, 5)
        elif B[2][4] == "0":
            p = (1, 5, 2, 4)
        elif B[1][4] == "0":
            p = (2, 4, 1, 4)
        elif B[2][5] == "0":
            p = (2, 4, 2, 5)
        else:
            p = rand_move()
    elif mx == cntr:
        if B[1][5] == "0":
            p = (0, 6, 1, 5)
        elif B[2][4] == "0":
            p = (1, 5, 2, 4)
        elif B[2][5] == "0":
            p = (2, 4, 2, 5)
        elif B[1][4] == "0":
            p = (2, 4, 1, 4)
        elif B[5][1] == "0":
            p = (6, 0, 5, 1)
        elif B[4][2] == "0":
            p = (5, 1, 4, 2)
        elif B[5][2] == "0":
            p = (4, 2, 5, 2)
        elif B[4][1] == "0":
            p = (4, 2, 4, 1)
        else:
            p = rand_move()    
    return p
```

```python
def phase23(A, dep, player):
    if dep == 0 :
        return (-1, -1, -1, -1, score(A, player))

    mx1 = -INF
    mxp1 = ()
    plist1 = all_move(A, player)
    if not plist1:
        p1 = (-1, -1, -1, -1)
        AA = next_board(A, p1, player)
        return p1 + (score(AA, player))
    for p1 in range(plist1):
        AA = next_board(A, p1, player)
        mn2 = INF
        plist2 = all_move(AA, "2" if player == "1" else "1")
        if not plist2:
            p2 = (-1, -1, -1, -1)
            AAA = next_board(AA, p2, "2" if player == "1" else "1")
            mn2 = score(AAA, 1)
        for p2 in plist2:
            AAA = next_board(AA, p2, "2" if player == "1" else "1")
            res = phase23(AAA, dep - 1)
            mn2 = min(mn2, res[4])
        if mx1 < mn2:
            mx1 = mn2
            mxp1 = p1
    return mxp1 + (mx1)
```
